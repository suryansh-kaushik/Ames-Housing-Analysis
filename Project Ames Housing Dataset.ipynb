```python
# for reporting back how much percent data is missing
def percent_missing(df):
    percent_nan= 100*df.isnull().sum() / len(df) 
    percent_nan= percent_nan[percent_nan>0].sort_values()
    
    return percent_nan
```

Model Formation

```python
# Predciting the sale price based on given data
X = final_df.drop('SalePrice',axis=1)
y = final_df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=101)
```

```python
#he dataset features has a variety of scales and units. For optimal regression performance, we are scaling the X features.
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet()
```

```python
param_grid={'alpha':[0.1,1.0,10,100],'l1_ratio':[0.1,1.0,10,100]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# verbose number a personal preference
grid_model = GridSearchCV(estimator=base_elastic_model,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error',
                          cv=5,
                          verbose=2)
```

```python
# fitting the model
grid_model.fit(scaled_X_train,y_train)
```

```python
grid_model.best_estimator_
#best combination found
```

```python
y_pred = grid_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```
